```python
import numpy as np
import pandas as pd

from scipy.stats import norm

def probability_between(mean, std_dev, lower_bound, upper_bound):
    # Calculate the cumulative probability for the lower and upper bounds
    # using the cumulative distribution function (cdf)
    lower_prob = norm.cdf(lower_bound, loc=mean, scale=std_dev)
    upper_prob = norm.cdf(upper_bound, loc=mean, scale=std_dev)

    # Calculate the probability between the two bounds
    probability = upper_prob - lower_prob
    return probability


def f(cgpa) :
  df = pd.DataFrame({'Cgpa' : [probability_between(cgpa, 1, 9.5, np.inf), probability_between(cgpa, 1, 8.5, 9.5), probability_between(cgpa, 1, 7.5, 8.5), probability_between(cgpa, 1, 6.5, 7.5), probability_between(cgpa, 1, 5.5, 6.5), probability_between(cgpa, 1, 4.5, 5.5),probability_between(cgpa, 1, 3.5, 4.5) ]})
  return df
```

```python
cgpa_data = f(8.6)
cgpa_data.head()
```
